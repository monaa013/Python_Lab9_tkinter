```python
#q1
import tkinter as tk
from tkinter import messagebox

# Function to submit the form
def submit_form():
    domain_name = domain_name_entry.get()
    pet_name = pet_name_entry.get()
    owner_name = owner_name_entry.get()
    pet_age = pet_age_entry.get()
    gender = gender_var.get()
    species = species_var.get()
    allergies = allergies_var.get()

    # Display the submitted details in a message box
    messagebox.showinfo("Form Submission", f"Domain Name: {domain_name}\nPet Name: {pet_name}\nOwner Name: {owner_name}\nPet Age: {pet_age}\nGender: {gender}\nSpecies: {species}\nAllergies: {allergies}")

# Create the main window
root = tk.Tk()
root.title("Domain Details Form")

# Create and place labels and entry widgets for domain name 
domain_name_label = tk.Label(root, text="Domain Name:")
domain_name_label.pack()
domain_name_entry = tk.Entry(root)
domain_name_entry.pack()

# Create and place labels and entry widgets for domain name
pet_name_label = tk.Label(root, text="Pet Name:")
pet_name_label.pack()
pet_name_entry = tk.Entry(root)
pet_name_entry.pack()

# Create and place labels and entry widgets for domain name
owner_name_label = tk.Label(root, text="Owner Name:")
owner_name_label.pack()
owner_name_entry = tk.Entry(root)
owner_name_entry.pack()

pet_age_label = tk.Label(root, text="Pet Age:")
pet_age_label.pack()
pet_age_entry = tk.Entry(root)
pet_age_entry.pack()

# Create and place radio buttons for gender selection
gender_label = tk.Label(root, text="Gender:")
gender_label.pack()
gender_var = tk.StringVar()
gender_var.set("Male")  # Default selection
male_radio = tk.Radiobutton(root, text="Male", variable=gender_var, value="Male")
female_radio = tk.Radiobutton(root, text="Female", variable=gender_var, value="Female")
male_radio.pack()
female_radio.pack()

# Create and place a dropdown menu for species selection
species_label = tk.Label(root, text="Species:")
species_label.pack()
species_var = tk.StringVar()
species_options = ["Dog(1)", "Cat(2)", "Rabbit(3)", "Hamster(4)", "Fish(5)", "Guinea Pig(6)", "Other(7)"]
species_dropdown = tk.OptionMenu(root, species_var, *species_options)
species_dropdown.pack()

# Create and place a dropdown menu for allergies selection
allergies_label = tk.Label(root, text="Allergies:")
allergies_label.pack()
allergies_var = tk.StringVar()
allergies_options = ["Pollens(1)", "Mold spores(2)", " Dust mites(3)", "Shed skin cells(4)", "Insect proteins(5)", "Flea saliva(6)", "Other"]
allergies_dropdown = tk.OptionMenu(root, allergies_var, *allergies_options)
allergies_dropdown.pack()

# Create a submit button
submit_button = tk.Button(root, text="Submit", command=submit_form)
submit_button.pack()

# Start the Tkinter main loop
root.mainloop()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate pets data
np.random.seed(0)
x = np.linspace(0, 10, 100)  # X-axis data
y = np.sin(x) + np.random.normal(0, 0.1, 100)  # Y-axis data

# Line Graph
plt.figure(figsize=(8, 4))
plt.plot(x, y, label="Line Graph", color="b")
plt.xlabel("Species")
plt.ylabel("Vaccines")
plt.title("Line Graph Between Species and Pets")
plt.legend()
plt.grid(True)

# Bar Graph
categories = ["1-3", "4-6", "7-9", "Above"]
values = np.random.randint(1, 10, len(categories))  # Random data for categories
plt.figure(figsize=(8, 4))
plt.bar(categories, values, color="g")
plt.xlabel("Pets Age")
plt.ylabel("Pets")
plt.title("Bar Graph of the Pets Age")

# Scatter Plot
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
plt.figure(figsize=(8, 4))
plt.scatter(x_scatter, y_scatter, label="Scatter Plot", color="b", marker="o")
plt.xlabel("Allergies")
plt.ylabel("Pets")
plt.title("Scatter Plot Between Allergies and Pets")
plt.legend()

# Correlation Heatmap
correlation_matrix = np.corrcoef(x, y)
plt.figure(figsize=(8, 4))
plt.imshow(correlation_matrix, cmap="viridis", origin="upper")
plt.colorbar()
plt.title("Correlation Heatmap Of Pets")

plt.tight_layout()
plt.show()
```
